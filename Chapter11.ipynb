## 함수

```python
# 함수정의
# 1) 2개의 정수를 전달받아 두 정수의 합을 구한 다음 그 결과를 반환하는 함수 정의
def add(x, y) : # 매개변수 : 2개의 정수를 저장할 변수
  # 함수의 기능 및 반환 값
  result = x + y # 매개변수에 저장된 2개의 정수의 합 저장
  return result # 합을 반환

# 2) 함수 호출(함수 사용)
# 함수명(인수,,,)
print(add(10,20)) # 인수(10,20) : add()함수의 매개변수 x, y에 전달
                  #  => x, y에 저장된 값을 합한 30을 반환
# print(add(10)) # 매개변수와 인수의 개수는 일치해야한다

# 반환값이 있는 함수는 값처럼 사용 => 변수 = 값
result = add(30,40) # add() 함수에 30, 40을 전달하여 반환 받은 값을 변수에 저장
print(result)
```

```python
# 함수 정의(매개변수, return이 없는 경우)
def say_hello() :
  print("hello")

# 함수 호출
say_hello() # 매개변수가 없으므로 전달할 인수가 없음
```

## 디폴트 매개변수

```python
# 정수를 4개 전달받아 그 합을 반환하는 함수 정의
def add_nums(num1, num2, num3, num4 = 40) :
  # 디폴트 매개변수는 반드시 맨 뒤의 매개변수부터 설정을 해야 한다
  return num1 + num2 + num3 + num4

# 함수 호출
print(add_nums(1,2,3))
```

## 가변 매개변수

```python
# 가변 매개변수 : 몇 개의 인수가 들어올지 모를 때 주로 사용
# 전달되는 인수들을 출력하는 함수 show() 함수 정의

def show(*args) :
  print(args, type(args))

#함수 호출
show(1, "python", 3)
show("a", 3.14)
```
