## 반복문

### 1) while문

```python
# <while문 예제>
# while문 활용 1~10 정수 출력

# 1 ~ 10 정수를 표현할 변수 생성
num = 1
while num <= 10 : # num에 저장된 값이 10보다 작거나 같을 동안 반복
  # 반복할 문장 작성
  print(num)
  num += 1 # num에 저장된 값 1씩 증가
```

### 2) for문

```python
# for 변수명 in 집합(컬렉션, 시퀀스 자료형, range()...) :
# (들여쓰기, tab) 반복실행할 문장

# 문자열(시퀀스 자료형)과 for문
for i in "hello" :
  print(i) # i에는 문자열의 요소 하나씩 순차적으로 저장됨
print("리스트와 for문")
for i in ["문자열", 1, 5.4, True] :
  print(i, type(i)) # i에는 리스트의 요소 하나씩 순차적으로 저장됨
```

### 3) range() 함수

```python
# range()함수와 for문

for i in range(10) :
  print(i) # 반복할 때마다 i에는 0 ~ 9 정수가 순차적으로 저장
print()

for i in range(1, 10) :
  print(i) # 반복할 때마다 i에는 1 ~ 9 정수가 순차적으로 저장
print()

for i in range(1, 10, 2) :
  print(i) # 반복할 때마다 i에는 1 ~ 9까지 2씩 증가하는 정수가 순차적으로 저장

# 반복횟수 지정
for i in range(3) : # 0 ~ 2 : 정수 3개
  print("hello")
```

### 4) 기타 제어문

```python
# 1) break문 : 반복문 강제 종료

# 1 ~ 10 범위 중 5까지만 출력
for i in range(1, 11) :
  print (i)
  # i 에 저장된 값이 5라는 반복문 종료
  if i == 5 :
    print("프로그램 종료")
    break # 반복문 강제 종료
```

```python
# 2) continue문 : 다음 반복 계속 진행
# 1 ~ 10 정수 중 5만 빼고 출력
for i in range(1,11) :
  if i == 5 :
    continue # 다음 반복으로 계속 진행
  else :
    print(i)
```

```python
# 3) pass문 : 조건문, 반복분에 실행내용을 정하지 않았을 때 임시로 작성

if 10 > 5 :
  pass # 조건문에서 넣어줄 실행문이 딱히 없을 경우

for i in range(3) :
    pass # 아직 작성할 내용을 정하지 않았으니 pass
```
