## 내장 메소드

```python
 # .count()
 str1 = "best of best"
 print(str1, type(str1))
 print(str1.count("best")) # 2. str1에서 'best'가 몇번 나오는지 알려줘
```

```python
# .find()
str1 = "best of best"
print(str1.find("best")) # str1에서 'best'가 처음 나온 위치
print(str1.find("worst")) # 찾는 문자열이 없는 경우 -1 반환

# index()
print(str1.index('best')) # 0. find()와 기능이 같다
# print(Str1.index("worst")) # 찾는 문자열이 없는 경우 오류 발생
```

```python
# .join()
li = ["h", "e", "l", "l", "o"]
print(li, type(li))
print("".join(li)) # 리스트의 요소들이 하나의 문자열로 합쳐짐
print("-".join(li)) # 리스트의 요소 사이마다 "-" 문자열이 추가되어 합쳐짐

di = {"a" : 10, "b" : 20}
print(di,type(di))
print("".join(di)) # 딕셔너리는 키만 연결됨
```

```python
# .split()
data = "Hello World"
print(data, type(data))

print(data.split(), type(data.split()))
# 공백 기준으로 문자열을 분리 후 리스트에 저장한 결과 반환

print(data.split("o"))
# "o"를 기준으로 문자열을 분리 후 리스트에 저장한 결과 반환
```

```python
# .replace()
data = "Life is too short"
print(data)
print(data.replace("short", "long"))
# data 문자열에서 "short"를 "long"으로 변환
```

```python
# 불필요한 문자열 제거 메소드
str1 = "*****1**2**3*****"

print(str1.strip("*"))
# str1에서 양쪽 끝에 있는 "*"을 제거
print(str1.lstrip("*"))
# str1에서 왼쪽 끝에 있는 "*"을 제거
print(str1.rstrip("*"))
# str1에서 오른쪽 끝에 있는 "*"을 제거
```
