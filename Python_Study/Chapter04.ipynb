## 형변환

```python
# 정수형 형변환 함수 : int()
# 1) 실수 -> 정수
num1 = 15.19
print(num1, type(num1))
print(int(num1), type(int(num1))) # 소수점은 버리고 정수로 변환됨

# 2) 문자열 -> 정수
print("15", type("15")) # 문자열
print(int("15"), type(int("15"))) # 정수 : "15" -> 15
# print(int("a")) 문자열 자료형을 정수로 변환할 때는 변환 가능한 값을 사용해야함

# 3) 논리형 -> 정수
print(True, type(True)) # 논리형
print(int(True), type(int(True))) #  True -> 1
print(int(False), type(int(False))) # False -> 0
```

```python
# 실수형 형변환 함수 : float()

# 1) 정수 -> 실수
print(float(1), type(float(1))) # 1 -> 1.0

# 2) 문자열 -> 실수
print(float("3.14"), type(float("3.14"))) # "3.14" -> 3.14

# 3) 논리형 -> 실수
print(float(True), type(float(True))) # True -> 1.0
print(float(False), type(float(False))) # False -> 0.0
```

```python
# 논리형 형변환 함수 : bool()
# True로 변환되는 값 : False로 변환되는 것을 제외한 모든 값
# False로 변환되는 값 : 0, 0.0, "", '', (), {}, [] => 값이 비어있는 것들

# 1)  숫자형 -> 논리형
print(bool(0), type(bool(0)))
print(bool(3.14), type(bool(3.14)))

# 2) 문자열 -> 논리형
print(bool(""), type(bool("")))
print(bool("abc"), type(bool("abc")))

# 3) 비어있는 괄호
print(bool([]), type(bool([])))
print(bool({}), type(bool({})))
```

```python
# 문자열 형변환

# 1) 정수 -> 문자열
print(str(10), type(str(10))) # 10 -> "10"

# 2) 실수 -> 문자열
print(str(3.14), type(str(3.14))) # 3.14 -> "3.14"

# 3) 논리형 -> 문자열
print(str(True), type(str(True))) # True -> "True"
print(str(False), type(str(False))) # False -> "False"
```

## 시퀀스 자료형

```python
# 1) 인덱싱 : 시퀀스 자료형에서 요소 1개만 추출,  인덱스 번호 사용
#       01234 앞에서부터 시작
data = "hello"
#      -54321 뒤에서부터 시작

print(data, type(data)) # 문자열(시퀀스 자료형)
print(data[0]) # idx 번호 0번째 위치에 있는 문자 추출
print(data[-5]) # idx 번호 -5번째 위치에 있는 문자 추출
```

```python
# 2) 슬라이싱 : 한 문자 이상 구성된 단어나 문자를 추출, 인덱스 번호 사용
#  변수명[start:stop:step]
# start : 시작 인덱스 번호
# stop : 종료 인덱스 번호
# step : 간격 (기본값 1)

print(data[0:2:1]) # idx번호 0~1까지 문자열 슬라이싱
print(data[0:2]) # 간격을 1로 설정하여 슬라이싱 할 때는 생략 가능
print(data[:2]) # idx 0부터 시작하면 start는 생략 가능
print(data[1:5]) # idx 1 ~ 4까지 문자열 슬라이싱
print(data[1:]) # idx 1 ~ 4까지 문자열 슬라이싱
print(data[:]) # idx 0 ~ 슬라이싱
print(data[::2]) # idx 0 ~ 4까지 간격 2로 슬라이싱
```

```python
# 3) 시퀀스 연산자

print("hi~" + "hello~") # 문자열 + 문자열 => 문자열 연결
print("hi" * 3) # 문자열 * 정수 => 문자열 반복
```
