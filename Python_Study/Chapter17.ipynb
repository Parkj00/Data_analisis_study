## 클래스

```python
# class 클래스명 :
#   클래스의 속성 : 데이터 속성, 메소드

class Person :
    # 사람의 이름을 정의하는 메소드
    def get_name(self, name) :
      self.name = name # self.데이터 속성 = 값
      # Person클래스의 name 데이터 속성에 매개변수 name에 저장된 인수 저장
    # 사람의 나이를 정의하는 메소드
    def get_age(self,age) :
      self.age = age

    # 사람이 자신을 소개하는 메소드
    def introduce(self) :
      print("제 이름은 ", self.name, "입니다")
      print("나이는", self.age, "살 입니다")
```

```python
# 클래스 사용 : 객체화
# 객체명 = 클래스명()

pr1 = Person() # Person의 pr1객체 생성

# 속성 접근
# 1) 데이터 속성 접근
pr1.name = "둘리" # pr1의 name 데이터 속성에 "둘리" 저장
print(pr1.name)

# 2) 메소드 접근
pr1.get_age(10) # pr1의 get_age() 메소드 호출
                # pr1.age = age : 데이터 속성 age에 매개변수 age에 저장된 인수 저장
print(pr1.age)

pr1.introduce()


# 두 번째 객체 생성
pr2 = Person()

pr2.name = "도우너" # pr2의 데이터 속성에 값 저장
pr2.age = 12

pr2.introduce()
```

## 생성자

```python
# 생성자가 추가된 Person 클래스 선언
class Person :
    # 생성자  : name, age를 객체가 생성될 때 초기화
  def __init__(self, name, age) :
      self.name = name
      self.age = age

  # 메소드
  def introduce(self) :
        print("제 이름은", self.name, "입니다")
        print("나이는", self.age, "살 입니다")


# 객체화
p1 = Person("둘리", 10) # 생성자는 객체를 생성할 때 실행
p2 = Person("도우너", 11)

p1.introduce()
p2.introduce()
```
