## 1. 세트

```python
set1 = {"사과", "참외", "파인애플", "멜론"}
print(set1, type(set1)) # 인덱스 번호를 사용할 수 없는 비시퀀스 자료형

# 세트의 메소드

# .add()
set1.add("포도")
print(set1) # set에 "포도"가 추가되어 있음
set1.add("사과")
print(set1) # set는 중복된 값을 가질 수 없음

# .remove(값)
set1.remove("참외")
print(set1) # "참외" 삭제
# set1.remove("수박") 없는 값을 삭제하려고 하면 오류 발생

# .discard(값)
set1.discard("포도")
print(set1) # 포도 삭제
set1.discard("수박") # 없는 값을 삭제하려고 해도 오류 발생하지 않음

# .pop() : 세트의 요소를 하나 삭제
set1.pop()
print(set1)
set1.pop()
print(set1)
```

### 세트의 집합 연산

```python
# 세트 생성
se1 = {1, 2, 3, 4, 5}
se2 = {3, 4, 5, 6, 7}

# 교집합
print(se1 & se2) # 2개의 세트의 공통 요소
inter_set = se1.intersection(se2)
print(inter_set)

# 합집합
print(se1 | se2)
union_set = se1.union(se2)
print(union_set)

# 차집합
print(se1 - se2)
di_set = se1.difference(se2)
print(di_set)
```

### 딕셔너리

```python
# 딕셔너리 생성
di = {"A" : 1, "B" : 2}
print(di, type(di))

# 딕셔너리의 key로 value에 접근
print(di["A"])

# 딕셔너리의 요소 수정
# 딕셔너리[key] = value(수정할 값)
di["A"] = 10
print(di) # "A" 의 value 10으로 변경됨

# 딕셔너리 메소드

# .get(key)
print(di.get("B")) # "B"의 value 2 반환

# .pop(key)
di.pop("B")
print(di) # "B"와 해당하는 value값 2 삭제

# 딕셔너리 요소 추가
# 딕셔너리[key] = value
# 기존에 없는 key 작성
di["C"] = 30
print(di) # key는 C 해당하는 value 30 추가됨
```

```python
print(di)

# .keys()
print(di.keys())

for i in di.keys() :
    print(i) # key를 하나씩 가져옴

# .values()
print(di.values())

for i in di.values() :
    print(i) # value를 하나씩 가져옴

# .items()
print(di.items())

for i in di.items() :
    print(i, type(i)) # (key, value)

for i in di.items() :
    # 튜플의 언패킹 활용
    key, value = i # ("A", 10)
    print(key, value)

for key, value in di.items() :
    print(key, value)
```
