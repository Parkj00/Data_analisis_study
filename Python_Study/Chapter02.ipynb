```python
# 출력 : 컴퓨터가 사람에게 전달하는 것
# print() : ()안에 있는 내용을 화면에 출력

# 출력함수의 2가지 옵션
# (1) end 옵션 : print문의 마지막 출력 설정 (기본값 : "\n"(줄바꿈))
# (2) sep 옵션 : print문의 출력문들 사이에 해당하는 내용을 넣을 수 있다

# "큰 따옴표", '작은 따옴표'
# 컴퓨터가 명령어가 아닌 물자열 값임을 구분할 수 있게 사용함
```

## 1) 출력함수

```python
1 # print() : () 소괄호 안에 있는 값을 화면에 출력

print("안녕하세요")
print(10)
print(10+20)
```

```python
1 # 1) end 옵션 : 출력 내용을 어떻게 끝낼 것인지 설정(기본값 : \n)
print("반갑습니다", end = "\n")
# \n : 줄바꿈
print("오늘은 2회차 수업입니다", end = "!!")
print("",end ="\n")
print("앞으로 잘 부탁드려요")
```

```python
# 2) sep : 구분자 옵션. 출력 요소 사이마다 어떤 내용을 넣을 건지 설정
print("오늘은", 2, "회차 수업입니다")
print("오늘은", 2, "회차 수업입니다", sep = "~~", end = "\n")

# 연습
print("오늘은", 8.6 ,"일입니다", sep = "ㅁ")
```

## 2) 작은 따옴표(''), 큰 따옴표 ("")

```python
print("문자열은 작은 따옴표나 큰 따옴표를 사용합니다")
print('문자열은 작은 따옴표나 큰 따옴표를 사용합니다')
```

```python
print("4+3") # 따옴표 안에 있으면 문자열
print(4+3) # 숫자 + 숫자 -> 덧셈 연산
print("4"+"3") # 문자열 + 문자열 -> 문자열 연결
```
