## 클래스의 속성

```python
class Student :

    school_name = "Python학교" # 클래스 변수, 모든 객체들이 공유하는 값을 저장

    # 생성자 : 학생의 이름과 나이를 초기화
    def __init__(self, name, age) :
        self.name = name # 인스턴스 변수, 객체들마다 다른 값을 저장
        self.age = age

        # 학생의 이름과 나이를 출력하는 메소드(인스턴스 메소드)
    def print_info(self) :
        print("학생의 이름 : ", self.name)
        print("학생의 나이 : ", self.age)

    # 학생들의 학교명을 출력하는 메소드
    @classmethod
    def print_school(cls) : # 클래스 메소드
        print("학생들은 ", cls.school_name, "에 다닌다!") # 클래스 변수 사용

    # 등교하기 위한 교통수단을 출력하는 메소드(정적 메소드)
    @staticmethod
    def print_trans() : # 정적 메소드
        print("학생들은 버스를 타고 등교한다!")


```

```python
# Student 클래스 객체 생성
st1 = Student("둘리", 12)

print("====데이터 속성====")

# 인스턴스 변수 호출 : 객체를 통해서만 접근 가능
print(st1.name)
print(st1.age)

# 클래스 변수 호출 : 객체와 클래스를 통해 접근 가능
print(st1.school_name)
print(Student.school_name)


print("====메소드 호출====")

# 인스턴스 메소드 호출 : 객체를 통해서만 호출
st1.print_info()

# 클래스 메소드 호출 : 객체와 클래스를 통해 호출 가능
st1.print_school()
Student.print_school()

# 정적 메소드 호출 : 객체와 클래스를 통해 호출 가능
st1.print_trans()
Student.print_trans()
```
