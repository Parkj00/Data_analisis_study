## 입력함수

```python
# 입력함수 : input()
# input() # 입력받는 창이 생성됨
input("입력해주세요 : ") # 입력 메세지 출력하면서 입력 받는 창이 생성됨
```

```python
name = input("이름 입력 : ") # 입력받은 값을 변수에 저장
print(name, type(name)) # 입력받은 값 출력. 타입은 문자열(str)

# 다른 자료형으로 입력받을 때는 형변환 함수 사용
age = int(input("나이 입력 : ")) # 입력받은 값을 int()함수를 사용하여 정수로 형변환
print(age, type(age))

# 실수 입력받기
number = float(input("실수를 입력해주세요 : ")) # 입력받은 값을 float()함수를 사용하여 실수로 형변환
print(number, type(number))
```

## 이스케이프 문자(제어 문자)

```python
# \", \" : 문자열 안에서 따옴표 표현
print("저는 \"박주영\"입니다.")

# \n : 줄바꿈
print("안녕\n하세요")

# \t : 일정 간격 띄우기, tab키(들여쓰기)
print("안녕\t하세요")

# \\ : 문자열 안에서 \를 표현
print("이스케이프 문자는 \\로 시작")
```
