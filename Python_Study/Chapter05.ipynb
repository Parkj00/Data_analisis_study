## 연산자

```python
# 최우선 연산자
print(10+2*3) #16
#     10 + 6
print((10+2)*3) #36
#       12  *3
```

```python
# 단항 연산자
# 부호표시 : +(양수), -(음수)
print(+10) # 양수
print(-10) # 음수
```

```python
# 산술 연산자

# 1) +
print(10+20) # 30
print(3.2+10) # 13.2

# 2) -
print(20-10) # 10
print(10-5.5) # 4.5

# 3) *
print(10*3) # 30
print(10*2.5) # 25.0

# 4) / => 항상 float으로 나옴
print(10/2) # 5.0

# 5) // => 몫 연산자
print(10//3) # 3 <= 10을 3으로 나눈 몫

# 6) % => 나머지 연산자
print(10%3) # 1 <= 10을 3으로 나눈 나머지

# 7) ** => 거듭제곱 연산자
print(5**3) # 5 * 5 * 5 = 125

# +) 나머지 연산자(짝수, 홀수 구분)
print(8 % 2) # 0. 짝수는 2로 나누면 나머지가 0
print(9 % 2) # 1. 홀수는 2로 나누면 나머지가 1
```

```python
# 관계 연산자 (결과는 bool로 나옴)
print(10 > 3) # 10이 3보다 크니? True
print(5 <= 10) # 5가 10보다 작거나 같니? True
print(10 >= 2) # 10이 2보다 크거나 같니 ? True
print(10 == 10) # 10이 10과 같니? True
print(10 != 10) # 10이 10과 같지 않니? False

# 문자열 비교
print("안녕" == "안녕") # True
print("hello" != "bye") # True
```

```python
# 논리 연산자

# 1) and : 모두 True면 True
print(True and True) # True
print(True and False) # False
print(False and True) # False
print(False and False) # False

# 2) or : 하나라도 True면 True
print(True or True) # True
print(True or False) # True
print(False or True) # True
print(False or False) # False

# 3) not : True -> False, False -> True
print(not True) # False
print(not False) # True
```
