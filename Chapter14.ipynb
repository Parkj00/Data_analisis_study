## 내장함수

### 1) 숫자 내장함수

```python
# abs() : 절대값 구하기

print(abs(-10)) # -10의 절대값 -> 10
print(abs(10)) # 10의 절대값 -> 10
```

```python
# divmod() : (몫, 나머지)를 tuple로 반환
print(divmod(10,3), type(divmod(10,3))) # (3, 1)
print(10 // 3) # 3. 몫
print(10 % 3) # 1. 나머지
```

```python
# max(), min(), sum()
li = [1, 3, 5, 7, 9]
print(max(li)) # li에서 가장 큰 값
print(min(li)) # li에서 가장 작은 값
print(sum(li)) # li의 요소들의 합
```

```python
# pow() : 거듭제곱
print(10 ** 2)
print(pow(10, 2)) # 10 ** 2와 같다
```

```python
# round() : 반올림
print(round(1.6)) # 2
print(round(2.32))
print(round(3.142, 2)) # 3.14 소수점 2자리로 반올림
```

### 2) 시퀀스 내장함수

```python
# enumerate() : 리스트와 함께 사용, (idx, 요소)
li = ["A", "B", "C"]

for i in enumerate(li) :
    print(i, type(i)) # (idx, 값)

for idx, element in enumerate(li) : # 튜플의 언패킹
    print(idx, element)
```

```python
# len()
str1 = "hello"
li1 = [1, 2, 3, 4]
di1 = {"a" : 1, "b" : 2}

print(len(str1)) # 5개의 요소를 가지는 시퀀스 자료형
print(len(li1)) # 4
print(len(di1)) # 2, 딕셔너리는 key, value가 항상 한쌍의 1개의 값이기 때문에
```

```python
# sorted()
li2 = [5, 2, 7, 1, 8, 2, 3]
print(sorted(li2)) # 오름차순 결과
print(sorted(li2, reverse = True)) # 내림차순 결과
```

```python
# zip()

alpha = ["A", "B", "C", "D", "E"]
nums = [1, 2, 3, 4]

# for문과 zip() 사용
for i in zip(alpha, nums) :
    print(i, type(i))
    # 두 반복 가능 객체의 인덱스 번호가 일치하는 요소들끼리 묶어서 반환
    # zip() 함수를 사용하면 길이가 짧은 반복 가능 객체가 기준이 됨
```
