```python
# 삼항연산자
# 조건식이 참일 때 실행 if 조건식 else 조건식이 거짓일 때 실행

print("참" if True else "거짓") # 참
print("참" if False else "거짓") # 거짓

print("10은 3보다 크다" if 10 > 3 else "10은 3보다 작거나 같다")
# 삼항연산자의 결과 변수에 저장
result = "10은 3보다 크다" if 10 > 3 else "10은 3보다 작거나 같다"
# =(대입 연산자)는 연산자 중 우선순위가 가장 낮다 : 삼항연산자가 먼저 실행되고 대입연산자를 사용해 변수에 저장
print(result)
```

```python
# 복합대입연산자 (누적연산자)
num1 = 10 # 변수에 10 저장
num1 += 2 # num1에 저장된 값에 +2를 하여 다시 num1에 저장 => num1 = num1 + 2
print(num1) #12

num2 = 20
num2 *= 3 # num2에 저장된 값에 *3을 하여 다시 num2에 저장 => num2 = num2 * 3
print(num2) #60
```

## 연산자 우선순위

```python
'''
연산자 우선순위
최우선 연산자
단항 연산자
산술 연산자
관계 연산자
논리 연산자
삼항 연산자
대입 연산자
'''
```

```python
num = 10 + 3
# 대입  산술
# 산술 연산자가 대입 연산자보다 우선순위가 높다
# num = 13
print(num)

print(10 > 3 and 5 < 6) # True
#      True.      True. => True
# 관계 연산자가 우선순위가 더 높다

print(not 10 > 3)
#     not(True) => False
```
