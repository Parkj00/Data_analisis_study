## 변수

```python
# 변수 생성
# 변수명 = 값
name = "박주영"   # name이라는 저장공간에 문자열 "박주영" 대입(저장)
number = 10     # number라는 변수(저장공간)에 10을 저장

# 변수에 저장된 값 사용
print(name) # name이라는 변수에 저장된 "박주영" 출력
print(number) # number라는 변수에 저장된 10 출력
```

```python
# 변수 삭제
# del 변수명
number2 = 20
print(number2)

del number2 # number2라는 저장공간 삭제
print(number2) # name 'number2' is not defined
```

```python
# +) 변수에 저장된 값 수정
num = 30
print(num) # 30
# 기존변수명 = 수정할 값
num = 100
print(num) # 100
```

```python
# 변수명에는 띄어쓰기 불가능
'''
my age = 10 (x)
myage = 10 (o)
my_age = 10 (o) 팟홀표기법, 스네이크 표기법 (소문자로만 구성, 단어가 바뀔 때 _ 로 연결)
myAge = 10 (o) 카멜표기법 (소문자로 시작, 단어가 바뀔 때 첫 문자를 대문자로 작성)
MyAge - 10 (o) 파스칼표기법 (클래스명을 만들 때 주로 사용)
'''
# 특수문자 사용이 불가능하다 (단, _(언더바)만 사용가능)
#@$%@ = 10 오류발생

# 숫자로 시작 불가능하다
#1num = 10 오류발생
num1 = 10
n1um = 10

print (num1)
print (n1um)
```

## 기본 자료형

```python
# 1) 정수형(int)
print(10, type(10))

# 2) 실수형(float)
print(3.14, type(3.14))

# 3) 논리형(bool)
print(True, type(True))
print(False, type(False))

# 4) 문자열(str)
print("문자열", type("문자열"))
print('문자열', type('문자열'))
```
