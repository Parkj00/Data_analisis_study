## 조건문

### 1) if문

```python
# if 조건식 :
# (들여쓰기, tab) 조건식이 참일 때 실행

# <if문 예제>
# 사용자가 입력한 정수가 100보다 크다면 "100보다 크다!" 출력

num = int(input("정수 입력 : ")) # 정수를 입력받아 변수에 저장

# 입력받은 정수가 100보다 크다면 "100보다 크다!" 출력
if num > 100 :
    # num > 100 => True일 경우 실행
    print("100보다 크다!")
```

### 2) if - else문

```python
# <if - else문 예제>
# 사용자로부터 정수를 입력받아 입력받은 정수가 짝수라면 "짝수" 출력
# 그게 아니라면 "홀수" 출력

# 정수 입력받기
num = int(input("정수 입력 : "))

# num에 저장된 값이 짝수라면 "짝수" 아니라면 "홀수" 출력

if num % 2 == 0 :
  # num이 짝수인 경우 (짝수는 2로 나누면 나머지가 0)
  print("짝수")
else :
  # num이 홀수인 경우 (홀수는 2로 나누면 나머지가 1)
  print("홀수")
```

### 3) if - elif - else문

```python
# <if - elif - else문>
# 사용자로부터 정수를 입력받아 입력받은 정수가
# 20보다 크다면 "20보다 크다" 출력
# 그게 아니라 10 이상이거나 20보다 작거나 같다면 "10이상 20이하"
# 그게 아니라 10 미만이거나 5보다 크거나 같다면 "10미만 5이상입니다"
# 전부 아니라면 "5보다 작습니다"

num = int(input("정수 입력 : "))

if num > 20 :
    print("20보다 크다")
elif num >=10 :
    print("10이상 20이하 입니다")
elif num >= 5:
    print("10미만 5이상 입니다")
else :
    print("5보다 작습니다")
```
