## 컬렉션

### 1) 리스트

```python

li = [1, 1, 3.2, True, "a"] # 리스트 변수에 저장
#     0  1   2    3     4
print(li, type(li))

# 리스트는 시퀀스 자료형 => 인덱스 번호를 사용할 수 있다
```

```python
# 1) 인덱싱
print(li[0], type(li[0])) # 1, 1의 자료형 int
print(li[3], type(li[3])) # True, True의 자료형 bool

# 2) 슬라이싱
print(li[1:4], type(li[1:4])) # idx 1~3까지 슬라이싱, 타입은 list
```

### 2) 튜플

```python
tu = (1, 1, 2, 3.3, True, "a") # 중복값을 가질 수 있으며, 모든 자료형 허용

print(tu, type(tu))

# 튜플은 시퀀스 자료형
# 1) 인덱싱
print(tu[1],type(tu[1])) # 1, 정수형(int)
# 2) 슬라이싱
print(tu[:3], type(tu[:3])) # idx 0 ~ 2까지 슬라이싱, 타입은 튜플

# +) 튜플의 요소가 1개일 경우
tu2 = (2,) # 마지막에 ,(콤마) 작성
print(tu2, type(tu2))

# ** 튜플의 특징
# 1) 패킹 : 여러가지 값들을 ,(콤마)로 이어서 작성하면 튜플로 묶어준다
tu3 = 1, 2, 3
print(tu3, type(tu3))

# 2) 언패킹 : 튜플의 각 요소를 여러개의 변수에 할당
num1, num2, num3 = tu3
# num1, num2, num3 = (1, 2, 3)
print(num1, type(num1))
print(num2, type(num2))
print(num3, type(num3))
```

## 3) 세트

```python
set1 = {1, 1, 2, 2, 3}
print(set1, type(set1)) # 중복값을 가질 수 없음

# 세트는 비시퀀스 자료형 => 인덱스 번호를 사용할 수 없다
print(set1[0]) # 오류
```

## 4) 딕셔너리

```python
# 딕셔너리
#{key : value, key : value}
di1 = {"A" : 1, "B" : 2}
print(di1, type(di1))

# value에는 여러 자료형 허용
di2 = {"A반" : ["둘리", "도우너"], "B반" : "마이콜"}
print(di2, type(di2))

# key로 value 접근 가능(key를 인덱스 번호처럼 사용)
# 변수명[인덱스번호]
# 딕셔너리명[key]
print(di2["A반"], type(di2["A반"])) # "A반"에 해당하는 value, value의 타입
print(di2['B반'], type(di2['B반'])) # "B반"에 해당하는 value, value의 타입
```

```python
# +) 비어있는 컬렉션 만들기
# 1) 리스트 => [], list() 이용하여 비어있는 리스트 생성
li1 = []
print(li1, type(li1))
li2 = list()
print(li2, type(li2))

# 2) 튜플 => (), tuple() 이용하여 비어있는 튜플 생성
tu1 = ()
print(tu1, type(tu1))
tu2 = tuple()
print(tu2, type(tu2))

# 3) tpxm => set() 함수 사용
set1 = set()
print(set1, type(set1))

# 4) 딕셔너리 => {}, dict() 이용하여 비어있는 딕셔너리 생성
di1 = {}
print(di1, type(di1))
di2 = dict()
print(di2, type(di2))
```
